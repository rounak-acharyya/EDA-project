```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\Rounak\Downloads\car_price.csv"
df = pd.read_csv(file_path)

print("Dataset Head:")
display(df.head())

print("\nDataset Info:")
df.info()

print("Dataset Tail:")
display(df.tail())

print("\nData Types:")
print(df.dtypes)
```

```python
#Data Cleaning
missing_values = df.isnull().sum()
dup_values = df.duplicated().sum()
print("\nMissing values per column:")
display(missing_values)
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
# Univariate Analysis
print("\nDescriptive Statistics:")
display(df.describe())
```

```python
num_columns = df.select_dtypes(include=['int64']).columns
# Histograms
for col in num_columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], bins=100) 
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.show()
# Box plots
for col in num_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x=col)
    plt.title(f'Box plot of {col}')
    plt.show()
# Density plots
for col in num_columns:
    plt.figure(figsize=(8, 4))
    sns.kdeplot(df[col], fill=True)
    plt.title(f'Density plot of {col}')
    plt.show()
```

From the Univariate Analysis we can say > Most of the cars were manufactured between 2012 and 2018
                                        > The selling price of most cars ranges from 0.1e6(100,000) to 0.3e6(300,000)
                                        > Most of the cars have been driven between 10,000 km and 80,000 km

```python
# Bivariate Analysis
# Scatter plots to identify trends and potential correlations with car price
numeric_df = df.select_dtypes(include=['int64'])
target_column = 'Selling_Price'
independent_vars = numeric_df.columns.drop(target_column)

# Generating scatter plots for each numeric feature against the target variable ("Selling_Price")
for col in independent_vars:
    plt.figure(figsize=(8, 4))
    sns.scatterplot(data=df, x=col, y=target_column)
    plt.title(f'Scatter plot of {target_column} vs {col}')
    plt.xlabel(col)
    plt.ylabel(target_column)
    plt.show()
```

```python
# Checking correlations with car price
plt.figure(figsize=(12, 8))

numeric_df = df.select_dtypes(include=['int64'])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
# numeric_df.corr(): This calculates the correlation matrix for the numeric columns in the DataFrame
plt.title('Correlation Heatmap')
plt.show()
```

From the bivariate analysis we can say > The newer the car, the higher the selling price.
                                       > The fewer kilometers a car has been driven, the higher the selling price.

```python
#Visualization:
# Categorical Data Analysis
cat_columns = df.select_dtypes(include=['object']).columns
# 1. Bar Charts
for col in cat_columns:
    if col != "Model":  # Skip bar chart for "Model" column as it is unique for every car
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df, x=col)
        plt.title(f'Bar chart of {col}')
        plt.xticks(rotation=45) #to read the x labels
        plt.show()
```

```python
#Pie Charts
for col in cat_columns:
    if col != "Model":  # Skip pie chart for "Model" column
        plt.figure(figsize=(8, 8))
        df[col].value_counts().plot.pie(autopct='%1.1f%%') # df[col].value_counts(): This counts the occurrences of each unique value in the col column
# .plot.pie(): This creates a pie chart from the value counts. autopct='%1.1f%%': This argument formats the percentage display on the pie chart,
# showing each slices percentage to one decimal point
        plt.title(f'Pie chart of {col}')
        plt.ylabel('')  # to hide y label
        plt.show()
```

From the Visualizations we observe > Most of the cars are from Maruti and Hyundai.
                                   > Most of the cars run of petrol and disel.
                                   > Most of the sellers are individuals.
                                   > Most of the cars are manual.
                                   > Most of the sellers are first owners.
